```julia
mutable struct Object
    p # position
    v # velocity
    m # mass
end
```

```julia
dt = 0.1
g = [0, 0, -9.8]

function step!(obj::Object)
    obj.p += dt * obj.v
    obj.v += dt * g
end
```

```julia
obj = Object([0, 0, 100], [50, -10, 50], 0)
```

```julia
x = []
y = []
z = []
for i = 1:1000
    if obj.p[3] <= 0 
        break 
    end
    append!(x, obj.p[1])
    append!(y, obj.p[2])
    append!(z, obj.p[3])
    step!(obj)
end
```

```julia
plot(x, z)
```

```julia
plot3d(x, y, z)
```
