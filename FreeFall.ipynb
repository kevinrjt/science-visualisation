```julia
mutable struct Body
    p # position
    v # velocity
    m # mass
end
```

```julia
g = [0, 0, -9.8]
```

```julia
function step!(body::Body, dt)
    body.p += dt * body.v
    body.v += dt * g
end
```

```julia
using Plots
pyplot()
```

```julia
body = Body([0, 0, 100], [50, -10, 50], 0)
x = []
y = []
z = []
for i = 1:1000
    if body.p[3] <= 0 
        break 
    end
    append!(x, body.p[1])
    append!(y, body.p[2])
    append!(z, body.p[3])
    step!(body, 0.1)
end
```

```julia
plot(x, z)
```

```julia
plot3d(x, y, z)
```

```julia
body = Body([0, 0, 100], [50, -10, 50], 0)
anim = Animation()
plt = plot3d(1, xlim=(0,600), ylim=(-200,0), zlim=(0,300), title="Free Fall", marker=5)
for i = 1:1000
    if body.p[3] <= 0 
        break 
    end
    push!(plt, 1, body.p[1], body.p[2], body.p[3])
    frame(anim)
    step!(body, 0.2)
end
gif(anim, "./FreeFall.gif", fps=40, show_msg=false)
```
