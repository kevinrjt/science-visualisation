```julia
mutable struct Body
    p # position
    v # velocity
    m # mass
end
```

```julia
GM = 1;
```

```julia
centroid = [0, 0]
```

```julia
using LinearAlgebra
```

```julia
function step!(body::Body, dt)
    direction = centroid - body.p
    dist = norm(direction)
    a = (GM / dist^3) * direction
    body.p += dt * body.v
    body.v += dt * a
end
```

```julia
using Plots
pyplot()
```

```julia
body = Body([1, 0], [0, 1], 0)
x = []
y = []
for i = 1:10000
    append!(x, body.p[1])
    append!(y, body.p[2])
    step!(body, 0.001)
end
```

```julia
plot(x, y)
```
